```python
from pathlib import Path
from Llm.llm_loader import LLM
import pandas as pd
from dotenv import load_dotenv
load_dotenv()
```

```python
Answer_llm = LLM(model_name = "gpt-5-mini-2025-08-07", provider= "openai")
```

```python
df = pd.read_json('Data/MMQA/Synthesized_three_table_with_spider_db_id.json')

prompt_path = Path('Templates/few_shot/baseline_schema_linking.txt')
idx = 100

record = df.iloc[idx]
question = record['Question']
hint = record.get('Hint', 'No hint')
if pd.isna(hint) or not str(hint).strip():
    hint = 'No hint'

schema_path = Path('Data/spider/database') / record['Spider_db_id'] / 'schema description.csv'
schema_df = pd.read_csv(schema_path)
database_schema = schema_df.to_markdown()

prompt_template = prompt_path.read_text(encoding='utf-8').strip()
prompt = (
    prompt_template
    .replace('{DATABASE_SCHEMA}', database_schema)
    .replace('{QUESTION}', question)
    .replace('{HINT}', hint)
)

```

```python
res = Answer_llm.query(prompt)
# print(os.environ.get("OPENAI_API_KEY"))
```
